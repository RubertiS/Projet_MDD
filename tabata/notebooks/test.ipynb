```python
import h5py

def inspect_hdf5_file(hdf5_file):
    with h5py.File(hdf5_file, 'r') as f:
        print(f"Contenu du fichier {hdf5_file}:")
        for key in f.keys():
            print(f"  - {key}: {f[key].shape}")
        print("---- Contenu des groupes : ----")
        for key in f.keys():
            print(f"  {key}: {list(f[key])}")

# Utiliser cette fonction pour inspecter le fichier
inspect_hdf5_file("data/Aircrafts/Aircraft_01.h5")  # Remplace par le chemin correct
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from tabata import Selector, Opset

# Charger les données d'un fichier HDF5 réel
def load_signals_from_hdf5(hdf5_file):
    opset = Opset(hdf5_file)
    print(opset)  # Informations sur le fichier
    return opset

# Détecter automatiquement des instants (par exemple, des pics locaux)
def auto_select_instants(selector, threshold=0.5):
    for df in selector:
        colname = df.columns[0]
        #print(f"Traitement du signal: {colname}")
        
        # Détecter des pics dans le signal
        y = df[colname].values
        peaks, _ = find_peaks(y, height=threshold)
        
        # Sélectionner le premier pic pour l'alignement
        if len(peaks) > 0:
            selector.selected[selector.sigpos] = peaks[0]
            selector.variables.add(colname)

# Alignement des signaux en fonction des instants sélectionnés
def align_signals(selector):
    aligned_signals = []
    for df in selector:
        colname = df.columns[0]
        #print(f"Alignement du signal: {colname}")
        
        # Obtenir l'instant clé détecté
        instant = selector.selected.get(selector.sigpos, 0)
        
        # Décaler le signal pour aligner l'instant clé
        aligned_signal = df.shift(-instant).dropna()
        aligned_signals.append(aligned_signal)
    
    return aligned_signals

# Visualisation des signaux alignés
def plot_aligned_signals(aligned_signals):
    plt.figure(figsize=(10, 6))
    for i, signal in enumerate(aligned_signals):
        plt.plot(signal.index, signal.iloc[:, 0], label=f"Signal {i+1}")
    plt.legend()
    plt.title("Signaux alignés")
    plt.xlabel("Temps")
    plt.ylabel("Amplitude")
    plt.show()

# Main
if __name__ == "__main__":
    # Charger les données depuis le fichier HDF5
    hdf5_file = "data/Aircrafts/Aircraft_01.h5"  # Remplace par le chemin correct
    opset = load_signals_from_hdf5(hdf5_file)
    
    # Initialiser un Selector pour détecter et aligner les instants
    selector = Selector(hdf5_file)
    
    # Sélectionner automatiquement les instants d'intérêt
    auto_select_instants(selector, threshold=0.5)
    
    # Aligner les signaux
    aligned_signals = align_signals(selector)
    
    # Visualiser les signaux alignés
    plot_aligned_signals(aligned_signals)
```
